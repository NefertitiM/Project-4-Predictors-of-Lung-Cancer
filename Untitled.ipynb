```python
import pandas as pd
import matplotlib.pyplot as plt

# Reading the CSV file from the resource folder
file_path = 'resources/Incidence_US_by_State_All_Races_Lung_&_Bronchus_(Both_Sexes)_2017_-_2021.csv'

df = pd.read_csv(file_path)

# Abbreviating the state names for better visualization
state_abbreviations = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DE", "DC", "FL", "GA", "HI", "ID", "IA", 
                       "KS", "KY", "LA", "ME", "MD", "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", 
                       "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", "SD", "TN", "TX", "UT", 
                       "VT", "VA", "WA", "WV", "WI", "WY"]

# Updating the state column in the DataFrame to use abbreviations
df["State"] = state_abbreviations

# Calculating the number of lung cancer cases caused by smoking
df["Lung Cancer from Smoking (Count)"] = (df["Percent of Smokers by State Population"] / 100) * df["Count of population with lung cancer"]

# Plotting both lung cancer cases caused by smoking and the total population by state with a secondary Y-axis for lung cancer cases
fig, ax1 = plt.subplots(figsize=(18,10))

# Plot 1: Population by state on primary Y-axis
ax1.bar(df["State"], df["popuation"], label="Population", color="skyblue", alpha=0.6)
ax1.set_xlabel("States")
ax1.set_ylabel("Population", color="blue")
ax1.tick_params(axis='y', labelcolor="blue")

# Create a second Y-axis for lung cancer cases
ax2 = ax1.twinx()
lung_cancer_from_smoking_relative = df["Lung Cancer from Smoking (Count)"]
ax2.bar(df["State"], lung_cancer_from_smoking_relative, label="Lung Cancer from Smoking (Count)", color="red", alpha=0.6)
ax2.set_ylabel("Lung Cancer Cases from Smoking (Count)", color="red")
ax2.tick_params(axis='y', labelcolor="red")

# Adding title
plt.title("Population and Lung Cancer Cases from Smoking by State")

# Show the plot
fig.tight_layout()
plt.show()

```
